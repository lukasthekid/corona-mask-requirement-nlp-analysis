```python
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
import pandas as pd
import gensim
import gensim.corpora as corpora
from wordcloud import WordCloud
from pprint import pprint
import os

import pyLDAvis.gensim_models
import pickle 
import pyLDAvis

import matplotlib.pyplot as plt
%matplotlib inline
```

Word embedding is a technique that represents words as numerical vectors in a low-dimensional space. These vectors capture the meanings of the words and the relationships between them, and can be used as input to machine learning models. Word embedding is particularly useful when you want to capture the semantics of the words and how they relate to each other.

TF-IDF, on the other hand, is a technique that represents the importance of a word in a document by considering its frequency in the document and the rarity of the word across all the documents in a corpus. TF-IDF is often used as a weighting scheme in information retrieval and text mining tasks, to give more weight to terms that are more important to a particular document.

In general, you should use word embedding when you want to capture the semantics of the words and their relationships to each other, and use TF-IDF when you want to identify the important words in a document or group of documents.

Which technique you should use will depend on the specific problem you are trying to solve and the type of data you have. If you have a large corpus of text data and you want to identify the main topics that are discussed in the documents, you may want to use a topic modeling algorithm such as Latent Dirichlet Allocation (LDA), which uses word frequencies and co-occurrences to identify the topics. If you have a smaller dataset and you want to classify the documents based on their content, you may want to use a machine learning model such as a support vector machine (SVM) or a neural network, which can be trained using word embeddings as input.

```python
df = pd.read_csv('../../data/p_comment.csv')
df.head()
```

```python
# Join the different processed titles together.
long_string = ','.join(list(df['merged_text'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, random_state=6, contour_color='steelblue', width=800, height=400)
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
# Save the wordcloud as a PNG file
wordcloud.to_file("../wordcloud.png")
wordcloud.to_image()
```

```python
tokens = [[x for x in text.split() if x != '|' and x != '--']for text in df['merged_text']]
# Create Dictionary
id2word = corpora.Dictionary(tokens)
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in tokens]
# View
print(corpus[:1][0])
```

```python
# number of topics
num_topics = 5
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
    with open(str(num_topics)+'_lda', 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
# load the pre-prepared pyLDAvis data from disk
with open(str(num_topics)+'_lda', 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, str(5)+'_lda' +'.html')
LDAvis_prepared
```
